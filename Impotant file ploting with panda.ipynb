we can do this by tow argument plt.plot

```python
years = [ 2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
plt.plot(years, yield_apples)
```

Axis labels 
add lebels to the axes to show what each axis represnts using the plt.xlabel and plt.ylabel method 

```python
plt.plot(years, yield_apples)
plt.xlabel('year')
plt.ylabel('yield (tons per hectare)');
```

```python
#plotting multiple lines

years = range(2000, 2012)
yield_apples= [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.919, 0.926, 0.929, 0.923, 0.918,0.895 ]
yield_oranges = [0.562, 0.741, 0.330, 0.723, 0.218,0.895, 0.914, 0.719, 0.526, 0.331, 0.434, 0.223]
```

```python
plt.plot(years,yield_apples, years, yield_oranges)

plt.xlabel('years')
plt.ylabel('yield (tons per hectare)')

plt.show()
```

```python
plt.plot(years,yield_apples, years, yield_oranges)

plt.xlabel('years')
plt.ylabel('yield (tons per hectare)')


plt.title("crop yields in dc")
plt.legend(['Apples', 'oranges']);
```

```python
plt.plot(years,yield_apples, marker="o")
plt.plot(years, yield_oranges, marker="x")
plt.xlabel('years')
plt.ylabel('yield (tons per hectare)')


plt.title("crop yields in dc")
plt.legend(['Apples', 'oranges']);
```

```python
#modifing markers and lines
plt.plot(years,yield_apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, yield_oranges, marker="x", c='r', ls='--', lw=3, ms=10, alpha=.5)
plt.xlabel('years')
plt.ylabel('yield (tons per hectare)')


plt.title("crop yields in dc")
plt.legend(['Apples', 'oranges']);
```
