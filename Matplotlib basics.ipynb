```python
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
from matplotlib.ticker import MaxNLocator
```

```python
x = np.arange(0,10,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y,'ok',x,y,'g')


plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.show()
```

```python
fig_2 = plt.figure(2, figsize=(6.4, 4.8))

x_1 = fig_2.add_subplot(121)
y_2 = fig_2.add_subplot(122)

x_1.plot(x,y)
y_2.scatter(x,y)

plt.show()
```

```python
apl_price = [87.98, 67.87, 56.74, 35.24]
ms_price = [65.01, 64.35, 89.53, 84.53]
year = [2012, 2013, 2014, 2015]
```

```python
plt.plot(year, apl_price, 'ok', year, ms_price)

plt.xlabel('year')
plt.ylabel('stock price')

plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(20, 4.8))
```

```python
fig_1 = plt.figure(1, figsize=(10, 4.8))

chart_1 = fig_1.add_subplot(122)
chart_2 = fig_1.add_subplot(121)

chart_1.plot(year, apl_price)
chart_1.xaxis.set_major_locator(MaxNLocator(integer=True))

chart_2.scatter(year, ms_price)
chart_2.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.show()
```

```python
fig_3, axes = plt.subplots(2,2,figsize=(20, 5))
axes[0,1].scatter(year, apl_price)
axes[1,0].scatter(year, ms_price)
plt.show()
```
